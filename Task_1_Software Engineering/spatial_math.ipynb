```python
import numpy as np
from spatialmath import *
import math

import matplotlib.pyplot as plt
%matplotlib notebook 
```

```python
# my plot is to have an item that spirals up in circle along z axis. 
# At each time frame, the theta and height are updated by 0.1, when theta reached 2pi, it completes a circle/spiral
# The object itself is also rotating on z-axis by theta at each time frame

h = 0
theta = 0

while (theta <= 2 * math.pi):
    spinner = SE3(0 + math.cos(theta), 0 + math.sin(theta), 0 + h) * SE3.Rz(theta, 'rad')
    spinner.plot(frame='A', color='black', dims=[-10,10])
    theta += 0.1
    h += 0.1
```
